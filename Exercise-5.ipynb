### Step 1. Import the necessary libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv(url, delimiter="|")
users.head(10)
```

### Step 4. Discover what is the mean age per occupation

```python
mean_per_occ = users.groupby("occupation")["age"].mean()
mean_per_occ
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
from fractions import Fraction

by_occupation = users.groupby("occupation")
count_per_occupation = by_occupation["user_id"].count()
males_per_occupation = users.groupby("occupation")["gender"].apply(lambda x: (x == "M").sum())

# print(by_occupation.groups) # groups returns a dictionary, containing occupations as keys and a list of user IDs as values.
# for occupation in by_occupation.groups.keys():
#     ratio = Fraction(males_per_occupation[occupation]/count_per_occupation[occupation]).limit_denominator()
#     print(f"{occupation}: {ratio}")

output = [f"{occupation}: {Fraction(males_per_occupation[occupation] / count_per_occupation[occupation]).limit_denominator()}" 
                    for occupation in by_occupation.groups.keys()] # with list comprehension

output
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_age = by_occupation["age"].min()
max_age = by_occupation["age"].max()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
# combo = users.groupby(["occupation", "gender"])["age"].mean()
# combo.unstack()
combo = users.groupby(["occupation", "gender"]).aggregate({"age": "mean"}) # looks better
combo
```

### Step 8.  For each occupation present the percentage of women and men

```python
females_per_occupation = users.groupby("occupation")["gender"].apply(lambda x: (x == "F").sum())
pct_males = round((males_per_occupation/count_per_occupation) * 100, 2)
pct_females = round((females_per_occupation/count_per_occupation) * 100, 2)
print(pct_females)
```
